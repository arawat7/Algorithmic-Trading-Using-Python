```python
from pandas_datareader import data as web
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2018,1,1)
end = datetime.datetime(2020,1,31)
```

```python
AAPL = web.DataReader('AAPL','yahoo',start,end)
```

```python
AAPL.head()
```

```python
len(AAPL)
```

```python
AAPL.Close.tail(500).plot(figsize=(16,6))
```

```python
#Close MA 20 days
AAPL['Close MA 20 Days'] = AAPL['Close'].rolling(20).mean()

# Upper = 20MA + 2*Std(20)
AAPL['Upper'] = AAPL['Close MA 20 Days'] + 2*(AAPL['Close'].rolling(20).std())

# Lower = 20MA - 2*Std(20)
AAPL['Lower'] = AAPL['Close MA 20 Days'] - 2*(AAPL['Close'].rolling(20).std())
```

```python
AAPL.head()
```

```python
AAPL[['Close','Upper','Lower','Close MA 20 Days']].tail(300).plot(figsize=(16,8))
```
